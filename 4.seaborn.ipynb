# 🔗 Jointplot in Seaborn

---

## 🔹 What is a Jointplot?

A **Jointplot** is used to visualize the **relationship between two numerical variables**.

👉 It combines:
- 📊 Scatter plot (relationship)
- 📈 Histogram / KDE (distribution)

---

## 🔹 Why Use Jointplot?

- Analyze **bivariate relationships**
- Understand **distribution of each variable**
- Detect **correlation & patterns**
- Useful in **EDA (Exploratory Data Analysis)**

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.jointplot(x="total_bill", y="tip", data=df)
plt.show()
````

````python
# scatter jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.show()
````

````python
# scatter jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.show()
````

````python
# kde jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="kde")
plt.show()
````

````python
# hex jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex")
plt.show()
````

````python
# regression jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
plt.show()
````

# 🔗 Pairplot in Seaborn

---

## 🔹 What is a Pairplot?

A **Pairplot** is used to visualize **pairwise relationships between multiple numerical variables** in a dataset.

👉 It creates:
- 📊 Scatter plots (between every pair of variables)
- 📈 Histograms/KDE (on diagonal)

---

## 🔹 Why Use Pairplot?

- Analyze **multiple relationships at once**
- Detect **correlation patterns**
- Identify **outliers**
- Useful in **EDA (Exploratory Data Analysis)**

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.pairplot(df)
plt.show()
````

````python
# pairplot with hue
sns.pairplot(df, hue="sex")
plt.show()
````

````python
# pairplot with selected columns
sns.pairplot(df[["total_bill", "tip", "size"]])
plt.show()
````

````python
# pairplot with selected columns
sns.pairplot(df[["total_bill", "tip", "size"]])
plt.show()
````

````python
# kde diagonal
sns.pairplot(df, diag_kind="kde")
plt.show()
````

````python
# change plot kind
sns.pairplot(df, kind="reg")
plt.show()
````

````python
# customize pairplot
sns.pairplot(df, hue="sex", palette="coolwarm")
plt.show()
````

# 📊 Distribution Plots (Distplot / Histplot) in Seaborn

---

## 🔹 What is a Distribution Plot?

A **Distribution Plot** is used to visualize the **distribution of a numerical variable**.

👉 It helps to understand:
- Data spread  
- Shape of distribution  
- Skewness  
- Outliers  

---

## 🔹 Note

⚠️ `distplot` is **deprecated** in Seaborn  
👉 Use:
- `histplot` (recommended)
- `kdeplot`

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.histplot(df["total_bill"])
plt.show()
````

````python
# histogram with kde
sns.histplot(df["total_bill"], kde=True)
plt.show()
````

````python
# histogram with kde
sns.histplot(df["total_bill"], kde=True)
plt.show()
````

````python
# kde plot
sns.kdeplot(df["total_bill"])
plt.show()
````

````python
# distribution with hue
sns.histplot(data=df, x="total_bill", hue="sex", kde=True)
plt.show()
````

````python
# multiple kde plots
sns.kdeplot(data=df, x="total_bill", hue="sex", fill=True)
plt.show()
````
