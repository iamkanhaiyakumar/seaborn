# 🔗 Jointplot in Seaborn

---

## 🔹 What is a Jointplot?

A **Jointplot** is used to visualize the **relationship between two numerical variables**.

👉 It combines:
- 📊 Scatter plot (relationship)
- 📈 Histogram / KDE (distribution)

---

## 🔹 Why Use Jointplot?

- Analyze **bivariate relationships**
- Understand **distribution of each variable**
- Detect **correlation & patterns**
- Useful in **EDA (Exploratory Data Analysis)**

---

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.jointplot(x="total_bill", y="tip", data=df)
plt.show()
````

````python
# scatter jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.show()
````

````python
# scatter jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.show()
````

````python
# kde jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="kde")
plt.show()
````

````python
# hex jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex")
plt.show()
````

````python
# regression jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
plt.show()
````
