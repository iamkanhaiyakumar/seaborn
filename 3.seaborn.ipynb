# 🔥 Heatmap in Seaborn

---

## 🔹 What is a Heatmap?

A **Heatmap** is a data visualization technique that shows values using **color intensity**.

👉 It is mainly used to visualize:
- Correlation between variables
- Matrix-like data
- Patterns in data

---

## 🔹 Why Use Heatmap?

- Understand **feature relationships**
- Identify **strong/weak correlations**
- Detect **patterns & trends**
- Useful in **EDA (Exploratory Data Analysis)**

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# correlation matrix
corr = df.corr(numeric_only=True)

print(corr)
````

````python
# basic code
sns.heatmap(corr)
plt.show()
````

````python
# heatmap with annotations
sns.heatmap(corr, annot=True)
plt.show()
````

````python
# heatmap with colormap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
````

````python
# customize heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
````

````python
# heatmap with mask
import numpy as np

mask = np.triu(corr)

sns.heatmap(corr, annot=True, cmap="coolwarm", mask=mask)
plt.show()
````

````python
# try this
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis")
plt.show()
````

# 🔗 Clustermap in Seaborn

---

## 🔹 What is a Clustermap?

A **Clustermap** is a heatmap with **hierarchical clustering**.

👉 It groups:
- Similar rows together  
- Similar columns together  

📌 Includes a **dendrogram (tree structure)** to show clustering

---

## 🔹 Why Use Clustermap?

- Identify **hidden patterns**
- Group similar data points
- Visualize **clusters in dataset**
- Useful in **machine learning & EDA**

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
# loaddata
df = sns.load_dataset("iris")

# preview data
print(df.head())
````

````python
# remove categorical column
df_numeric = df.drop("species", axis=1)
````

````python
# basic code
sns.clustermap(df_numeric)
plt.show()
````

````python
# clustermap with colormap
sns.clustermap(df_numeric, cmap="coolwarm")
plt.show()
````

````python
# clustermap with annotations
sns.clustermap(df_numeric, cmap="viridis", figsize=(8, 8))
plt.show()
````

````python
# try this
sns.clustermap(df_numeric.corr(), cmap="coolwarm")
plt.show()
````
