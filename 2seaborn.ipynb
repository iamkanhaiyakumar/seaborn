# 🎻 Violin Plot in Seaborn

## 🔹 What is a Violin Plot?
A **Violin Plot** is used to visualize the distribution of numerical data across categories.

It combines:
- 📦 Boxplot (median, quartiles)
- 📊 KDE (Kernel Density Estimation)

👉 It shows both **distribution shape + summary statistics**

---

## 🔹 Why Use Violin Plot?
- Understand **data distribution**
- Detect **skewness**
- Compare **multiple categories**
- More informative than boxplot

---

## 🔹 Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt

🔹 Load Dataset

````python
df = sns.load_dataset("tips")
````

🔹 Basic Violin Plot

````python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
````

🔹 Violin Plot with Hue

````python
sns.violinplot(x="day", y="total_bill", hue="sex", data=df)
plt.show()
````

🔹 Split Violin Plot

````python
sns.violinplot(x="day", y="total_bill", hue="sex", data=df, split=True)
plt.show()
````

🔹 Horizontal Violin Plot

````python
sns.violinplot(x="total_bill", y="day", data=df)
plt.show()
````

🔹 Customize Violin Plot

````python
sns.violinplot(x="day", y="total_bill", data=df, palette="pastel")
plt.title("Total Bill Distribution by Day")
plt.show()
````

🔥 Key Components
Width → Data density
White dot → Median
Thick bar → Interquartile Range (IQR)
Thin lines → Range

# 📊 Barplots and Countplots in Seaborn

## 🔹 Introduction
Seaborn provides powerful categorical plots like:
- 📊 **Barplot** → shows aggregated values (mean, etc.)
- 🔢 **Countplot** → shows frequency/count of categories

---

# 🔸 1. Barplot

## 🔹 What is Barplot?
A **barplot** represents the **aggregate value** (by default: mean) of a numerical variable for each category.

👉 It answers: *“What is the average value per category?”*

---

## 🔹 Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
````

````python
# barplot with hue
sns.barplot(x="day", y="total_bill", hue="sex", data=df)
plt.show()
````

````python
# horizontal barplot
sns.barplot(x="total_bill", y="day", data=df)
plt.show()
````

````python
# customize barplot
sns.barplot(x="day", y="total_bill", data=df, palette="pastel")
plt.title("Average Total Bill by Day")
plt.show()
````

# 📊 Countplot in Seaborn

---

## 🔹 What is Countplot?

A **Countplot** is used to show the **frequency (count)** of observations in each category.

👉 It is similar to a **bar chart**, but instead of showing mean values, it shows **number of occurrences**.

---

## 🔹 Why Use Countplot?

- Understand **distribution of categorical data**
- Compare **counts across categories**
- Identify **imbalanced data**
- Useful in **EDA (Exploratory Data Analysis)**

---

## 🔹 Import Libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

````python
# loaddata
df = sns.load_dataset("tips")

# preview data
print(df.head())
````

````python
# basic code
sns.countplot(x="day", data=df)
plt.show()
````

````python
# countplot with hue
sns.countplot(x="day", hue="sex", data=df)
plt.show()
````

````python
# horizontal countplot
sns.countplot(y="day", data=df)
plt.show()
````

````python
# countplot with order
sns.countplot(x="day", data=df, order=["Thur", "Fri", "Sat", "Sun"])
plt.show()
````

````python
# customize countplot
sns.countplot(x="day", data=df, palette="coolwarm")
plt.title("Count of Customers by Day")
plt.show()
````

````python
# labels and title
sns.countplot(x="day", data=df)
plt.xlabel("Days")
plt.ylabel("Number of Customers")
plt.title("Customer Count per Day")
plt.show()
````

````python
# try this
sns.countplot(x="time", hue="sex", data=df)
plt.show()
````
