# 📊 Categorical Scatterplots: Stripplot & Swarmplot in Seaborn
## 🔹 What are Categorical Scatterplots?
### Used to visualize relationship between a categorical variable and a numeric variable
## Helps in seeing distribution + individual data points

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='day',y='tip', kind='strip', data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=tips, jitter=False)
```

```python
sns.catplot(x='day',y='tip', kind='swarm', data=tips)
```

```python
sns.catplot(x='day',y='tip', kind='swarm', hue='sex', data=tips)
```

# 🔹 What is a Boxplot?

## A boxplot (box-and-whisker plot) shows:

### Distribution of data
### Median
### Quartiles
### Outliers

👉 Very useful in EDA (Exploratory Data Analysis)

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.boxplot(x="total_bill", y="day", data=df)
plt.show()
```

```python
sns.boxplot(x="total_bill", y="day", data=df)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, palette="pastel")
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
sns.boxplot(x="day", y="tip", data=df)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
sns.stripplot(x="day", y="total_bill", data=df, color="black")
plt.show()
```
